### Задание 3

Построй 95% доверительный интервал для среднего времени приготовления пиццы в **секундах**. Интервал округли до целого
> Для расчета стандартного отклонения используй `numpy.std(ddof=1)`

```python
def confidence_int(data, z=0.95):
    # array = 1.0 * np.array(data)
    n = len(data)
    x = np.mean(data)
    s = np.std(data, ddof=1)
    dev = (s / np.sqrt(n))
    int1, int2 = st.norm.interval(confidence=z, loc=x, scale=dev)
    return round(int1), round(int2)
print(' 95% доверительный интервал для среднего времени приготовления пиццы в секундах:', confidence_int(df.cook_time, 0.95))
```

### Задание 7

Для более детальной оценки построй 95% доверительный интервал для разности двух долей.
**Ответ дай в формате:**\
95% доверительный интервал для разности двух долей: [0.99847, 9.99847]

Левая и правая границы доверительного интервала рассчитываются по формулам:
$$p_a - p_b \pm z * \sqrt{\frac{p_a(1 - p_a)}{n_a} + \frac{p_b(1 - p_b)}{n_b}}$$
Где

$p_a = \frac{n_{like(a)}}{n_{a}}$,

$p_b = \frac{n_{like(b)}}{n_{b}}$,

$z = 1-\frac{\alpha}{2}$ (используй scipy.stats.norm.ppf)

$n_a$ - общее количество кликов по баннеру а

$n_b$ - общее количество кликов по баннеру b

$n_{like(a)}$ - количество лайков по баннеру а

$n_{like(b)}$ - количество лайков по баннеру b

```python
import scipy.stats

def di_stats(banner_a, banner_b):
    z = scipy.stats.norm.ppf(1-0.05 /2)
    n_a = len(banner_a)
    n_b = len(banner_b)
    p_a = float(sum(banner_a)) / n_a
    p_b = float(sum(banner_b)) / n_b
    p_di1 = p_a - p_b - z * np.sqrt((p_a * (1 - p_a)) / n_a +(p_b * (1 - p_b)) / n_b)
    p_di2 = p_a - p_b + z * np.sqrt((p_a * (1 - p_a)) / n_a +(p_b * (1 - p_b)) / n_b)
    return p_di1, p_di2

print('95% доверительный интервал для разности двух долей:', di_stats(click['banner_a'], click['banner_b']))
```

### Задание 8

Проведи АБ-тест. Какой баннер лучше? H0 - Старый баннер или H1 - новый баннер.
В этом тебе поможем функция [z-критерия](code-samples/stats.py).

$$Z({X_1, X_2}) =  \frac{p_a - p_b}{\sqrt{P(1 - P)(\frac{1}{n_a} + \frac{1}{n_b})}}$$
$$P = \frac{p_a{n_a} + p_b{n_b}}{{n_a} + {n_b}} $$

$p_a, p_b = \frac{∑кликов}{n}$, n - общее количество кликов по баннеру

$n_a$ - общее количество кликов по баннеру а

$n_b$ - общее количество кликов по баннеру b

```python
def z_stat(sample1, sample2):
    n_a = len(sample1)
    n_b = len(sample2)

    p_a = float(sum(sample1)) / n_a
    p_b = float(sum(sample2)) / n_b 
    P = float(p_a * n_a + p_b * n_b) / (n_a + n_b)
  
    return (p_a - p_b) / np.sqrt(P * (1 - P) * (1. / n_a + 1. / n_b))
```

```python
print('Какой баннер лучше гипотеза (banner_a лучьше banner_b) отрицательный нет:', z_stat(click['banner_a'], click['banner_b']))
print('Какой баннер лучше гипотеза (banner_b лучьше banner_a) положительный да:', z_stat(click['banner_b'], click['banner_a']))
```

```python
import numpy as np
x = np.linspace(-10,10,100)
y1 = x ** 2
y2 = -x **2
y3 = np.log(x ** 2 + 1)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import spearmanr
from scipy.stats import pearsonr

fig = plt.figure(figsize = (18,5))
axis = fig.subplots(nrows=1, ncols=3)

# For Sine Function
rho1, ps1 = spearmanr(x, y1) 
p11, p21 = pearsonr(x, y1)
sper1 = 'Spearman correlation ={}, \nPearson correlation ={}, '.format(rho1.round(4), p11.round(4))
axis[0].plot(x, y1, '.')
axis[0].set_title(sper1)
axis[0].set_xlabel('x')
axis[0].set_ylabel('y')

# For Cosine Function
rho2, ps2 = spearmanr(x, y2)
p12, p22 = pearsonr(x, y2)
sper2 = 'Spearman correlation ={}, \nPearson correlation ={}, '.format(rho2.round(4), p12.round(4))
axis[1].plot(x, y2, '.')
axis[1].set_title(sper2)
axis[1].set_xlabel('x')
axis[1].set_ylabel('y')

# For Cosine Function
rho3, ps3 = spearmanr(x, y3)
p13, p23 = pearsonr(x, y3)
sper3 = 'Spearman correlation ={}, \nPearson correlation ={}, '.format(rho3.round(4), p13.round(4))
axis[2].plot(x, y3, '.')
axis[2].set_title(sper3)
axis[2].set_xlabel('x')
axis[2].set_ylabel('y')
```
